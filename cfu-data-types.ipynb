### Exercise 1: Calculate remaining salary

In this exercise, prompt the user to enter their personal information such as name, age, address, salary, and expenses, and then perform some calculations to determine how much money they have left after expenses. 

Use data type casting to ensure that the age, salary, and expenses are stored as integers or floats, as appropriate, and round the remaining money to one decimal. 

Use a boolean variable to indicate whether the remaining salary is greater than or equal to 500. 

Finally, you will use string formatting to print a message to the screen in the following format: 
"*name*, who is *age* years old, and lives in *address* has *remaining_salary* dollars left from her salary after expenses. It is *is_salary_good* that she has more than $500 left.", using the mentioned variables.

*Hint:* 
- *You can use the input() function to ask the user to enter their information. Beware that input() function returns a String.*
- *You can round the remaining salary to one decimal using the round() function.*
- *Use data type conversion functions. Recommended external resources: [Data type conversion](https://www.geeksforgeeks.org/type-conversion-in-python/)*

```python
# ask to enter the name (str)
# ask to enter the country (str)
# ask to enter the age (int)
# as to enter the salary (flaot)
# ask to enter the expenses (flaot)
# ---------------------------------------#
# Operation:
# remaining_salary =  subtract from salary the expenses and round it to 1 decimal
# compare the remaining_salary with 500 return boolean
# ----------------------------------------#
# Result
# print "*name*, who is *age* years old, and lives in *country* has *remaining_salary* -->
# dollars left from her salary after expenses. It is *is_salary_good* that she has more than $500 left.

# Take the input
user_name = input("Enter your name: ").strip()
user_country = input("Enter your contry: ").strip()
user_age = int(input("Enter your age: ").strip())
user_salary = round(float(input("Enter your salary: ").strip()), 1)
user_expenses = round(float(input("Enter your expenses: ").strip()), 1)

# Calculate the remaining salary and round it to 1 digit
remaining_salary = round(user_salary - user_expenses, 1)

# return bool if the remanining salary is > 500 or <500
if remaining_salary <= 500:
    is_salary_good = False
else:
    is_salary_good = True

print(
    f"{user_name} who is {user_age} years old, and lives in {user_country} has {remaining_salary} EUR left from the salary after expenses. It is {is_salary_good} that he has more than 500 EUR left "
)
```

```python
poem = """Some say the world will end in fire,
Some say in ice.
From what I’ve tasted of desire
I hold with those who favor fire.
But if it had to perish twice,
I think I know enough of hate
To say that for destruction ice
Is also great
And would suffice."""

text = "Python is awesome!"
punctuation = "'.,!?\""

mod_poem = []
lower_poem = []

for i in poem:
    if i.strip(punctuation):
        mod_poem.append(i)

for x in mod_poem:
    lower_poem.append(x.lower())


# print(lower_poem)

new_poem = "".join(lower_poem).split()
# print(new_poem)
# print(type(new_poem))
conca_poem = new_poem + list("".join(text).split())
print(conca_poem)
```
